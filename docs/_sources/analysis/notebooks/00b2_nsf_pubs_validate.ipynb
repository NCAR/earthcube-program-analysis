```python
for k in pubs.keys():
    for i in pubs[k]:
        if i.get('doi') and i.get('cr_meta') and i['cr_meta'].get('DOI').lower() != i['doi'].lower():
            print(f"[DOI mismatch at {k}]: crossref meta={i['cr_meta']['DOI']} ==> nsf_mapping.json='{i['doi']}'")
```
